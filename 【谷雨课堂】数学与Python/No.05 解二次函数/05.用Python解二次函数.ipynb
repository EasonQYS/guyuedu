```python
import math

#一元二次方程 算法函数
def quadratic(a, b, c):
    if a == 0:
        if b == 0:
            if c == 0:
                print('X可以是任意数')
            else:
                print('无解')
        else:
            x = -c / b
            print('方程有解：x=%.2f' % x)
    else:
        delta = b * b - 4 * a * c
        if delta > 0:
            x1 = (-b + math.sqrt(delta)) / a / 2
            x2 = (-b - math.sqrt(delta)) / a / 2
            print("解为x1=%.2f，x2=%.2f" % (x1, x2))
            return 
        if delta == 0:
            x1 = -b / a / 2
            x2 = x1
            print("解相同，x1=x2=%.2f" % (x1))
            return x1, x2         
        print("无解,因为,B方-4AC=%d" % delta)
        
        
 
```

```python
quadratic(10,23,0)
```
