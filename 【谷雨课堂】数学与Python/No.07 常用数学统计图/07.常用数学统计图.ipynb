# 用Python制作常用的统计表

## 导入常用函数库

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 柱状图（条形图）

```python
name_list = ['AA', 'BB', 'CC', 'DD']
num_list = [10, 100 , 50, 1]
plt.bar(range(len(num_list)), num_list, tick_label=name_list,color='rbg')
plt.show()
```

### 折线图（曲线图）

```python
name_list = ['1', '2', '3', '4', '5', '6']
num_list = [1000, 5000, 500, 2000, 6000, 30000]
plt.plot(name_list,num_list)
plt.show()
```

### 饼图（扇形统计图）

```python
name_list = ['A', 'B', 'C', 'D']
num_list = [40, 20, 30, 10]

plt.axes(aspect=1)
plt.pie(x=num_list, labels=name_list, autopct='%3.1f %%')
plt.show()
```
